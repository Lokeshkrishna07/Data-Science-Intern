```python
import pandas as pd

# Load the dataset
file_path = 'Customers.csv'
customers_data = pd.read_csv(file_path)

# Convert 'SignupDate' to datetime format
customers_data['SignupDate'] = pd.to_datetime(customers_data['SignupDate'])

# Summary statistics
summary_stats = customers_data.describe(include='all')

# Distribution by region
region_distribution = customers_data['Region'].value_counts()

# Signups per year
signup_trend = customers_data['SignupDate'].dt.year.value_counts().sort_index()

# Output results
print("Summary Statistics:")
print(summary_stats)
print("\nRegion Distribution:")
print(region_distribution)
print("\nSignup Trend by Year:")
print(signup_trend)
```

```python
import pandas as pd

# Load the dataset
file_path = 'Products.csv'  # Update this path with your actual file path
products_data = pd.read_csv(file_path)

# Display basic information about the dataset
print("Dataset Information:")
print(products_data.info())

# Display the first few rows of the dataset
print("\nFirst 5 Rows:")
print(products_data.head())

# Summary statistics for all columns
print("\nSummary Statistics:")
print(products_data.describe(include='all'))

# Category-wise distribution of products
print("\nCategory Distribution:")
category_distribution = products_data['Category'].value_counts()
print(category_distribution)

# Price analysis
print("\nPrice Analysis:")
price_summary = products_data['Price'].describe()
print(price_summary)

# Price distribution by category
print("\nPrice Distribution by Category:")
category_price_summary = products_data.groupby('Category')['Price'].describe()
print(category_price_summary)
```

```python
import pandas as pd

# Load the dataset
file_path = 'Transactions.csv'  # Update this with your file's path
transactions_data = pd.read_csv(file_path)

# Strip any leading or trailing spaces from column names
transactions_data.columns = transactions_data.columns.str.strip()

# Display basic dataset information
print("Dataset Information:")
print(transactions_data.info())

# Display the first few rows of the dataset
print("\nFirst 5 Rows:")
print(transactions_data.head())

# List all column names to identify the exact name of the 'TransactionAmount' column
print("\nColumns in the dataset:")
print(transactions_data.columns)

# Summary statistics for all columns
print("\nSummary Statistics:")
print(transactions_data.describe(include='all'))

# Check for missing values
print("\nMissing Values:")
print(transactions_data.isnull().sum())

# Unique customers and transactions
print("\nUnique Customers and Transactions:")
unique_customers = transactions_data['CustomerID'].nunique()
unique_transactions = transactions_data['TransactionID'].nunique()
print(f"Unique Customers: {unique_customers}")
print(f"Unique Transactions: {unique_transactions}")

# Transaction analysis: Total, mean, and count
print("\nTransaction Analysis:")

# Check if 'TransactionAmount' column exists
if 'TransactionAmount' in transactions_data.columns:
    transaction_summary = transactions_data['TransactionAmount'].describe()
    print(transaction_summary)
else:
    print("Error: 'TransactionAmount' column not found in the dataset.")

# Group transactions by customers (checking for 'CustomerID' and 'TransactionAmount')
if 'CustomerID' in transactions_data.columns and 'TransactionAmount' in transactions_data.columns:
    print("\nTransactions Grouped by Customers:")
    customer_transactions = transactions_data.groupby('CustomerID')['TransactionAmount'].sum().sort_values(ascending=False)
    print(customer_transactions.head())
else:
    print("Error: Either 'CustomerID' or 'TransactionAmount' column is missing.")

# Transactions by date
print("\nTransactions by Date:")
if 'TransactionDate' in transactions_data.columns and 'TransactionAmount' in transactions_data.columns:
    transactions_by_date = transactions_data.groupby('TransactionDate')['TransactionAmount'].sum()
    print(transactions_by_date.head())
else:
    print("Error: Either 'TransactionDate' or 'TransactionAmount' column is missing.")
```
